##TTS models for train dataset

###Piper TSS

```python
!pip install piper-tts -q

from piper import PiperVoice
from piper.download_voices import download_voice
from pathlib import Path
from IPython.display import Audio, display
import os, wave

piper_dir = Path("piper_models")
os.makedirs(piper_dir, exist_ok=True)
download_voice("da_DK-talesyntese-medium", piper_dir)
piper_voice = PiperVoice.load(piper_dir / "da_DK-talesyntese-medium.onnx")
print("✓ Piper model loadet!")

text = "Hej, dette er en test af Piper TTS på dansk."
filepath = "piper_test.wav"

with wave.open(filepath, "wb") as wav_file:
    piper_voice.synthesize_wav(text, wav_file)

print("Stemme: da_DK-talesyntese-medium")
display(Audio(filepath))
```

###OpenAI TTS1 and GPT-4o-mini-tts

```python
from openai import OpenAI
import os
from IPython.display import Audio, display

client = OpenAI(api_key="*****")

voices_tts1 = [
    "alloy", "ash", "coral", "echo",
    "fable", "nova", "onyx", "sage", "shimmer"
]

voices_gpt4o = [
    "ballad", "verse", "marin", "cedar"
]

text = "Hej, dette er en test af OpenAI TTS systemet."

os.makedirs("openai_samples", exist_ok=True)

for voice in voices_tts1:
    response = client.audio.speech.create(
        model="tts-1",
        voice=voice,
        input=text
    )
    filepath = f"openai_samples/{voice}.mp3"
    response.stream_to_file(filepath)
    print(f"✓ Gemt: {filepath}")

for voice in voices_gpt4o:
    response = client.audio.speech.create(
        model="gpt-4o-mini-tts",
        voice=voice,
        input=text
    )
    filepath = f"openai_samples/{voice}.mp3"
    response.stream_to_file(filepath)
    print(f"✓ Gemt: {filepath}")

folder = "openai_samples"

for filename in sorted(os.listdir(folder)):
    if filename.endswith(".mp3"):
        voice_name = filename.replace(".mp3", "")
        print(f"Stemme: {voice_name}")
        display(Audio(f"{folder}/{filename}"))


print("Færdig! Alle 13 OpenAI stemmer er genereret.")
```

```python
 import nest_asyncio

async def generer_edge(text, voice):
    communicate = edge_tts.Communicate(text, voice)
    await communicate.save("temp.mp3")
    os.system("ffmpeg -i temp.mp3 -ar 24000 -ac 1 temp.wav -y -loglevel quiet")

def generer_lyd(text, navn, voice, path, model=None):
    try:
        if navn.startswith("google"):
            data = {
                "input": {"text": text},
                "voice": {"languageCode": "da-DK", "name": voice},
                "audioConfig": {"audioEncoding": "MP3"}
            }
            for forsøg in range(5):
                result = requests.post(
                    f"https://texttospeech.googleapis.com/v1/text:synthesize?key={GOOGLE_API_KEY}",
                    json=data
                ).json()
                if "audioContent" in result:
                    with open("temp.mp3", "wb") as f:
                        f.write(base64.b64decode(result["audioContent"]))
                    os.system(f"ffmpeg -i temp.mp3 -ar 24000 -ac 1 {path} -y -loglevel quiet")
                    break
                import time
                time.sleep(60)

        elif navn.startswith("edge"):
            asyncio.get_event_loop().run_until_complete(generer_edge(text, voice))
            os.system(f"ffmpeg -i temp.mp3 -ar 24000 -ac 1 {path} -y -loglevel quiet")

        elif navn.startswith("openai"):
            response = client.audio.speech.create(model=model, voice=voice, input=text)
            response.stream_to_file("temp.mp3")
            os.system(f"ffmpeg -i temp.mp3 -ar 24000 -ac 1 {path} -y -loglevel quiet")

        elif navn == "piper":
            with wave.open(path, "wb") as wav_file:
                piper_voice.synthesize_wav(text, wav_file)

        if os.path.exists(path):
            audio, sr = sf.read(path)
            return len(audio) / sr
    except Exception as e:
        print(f"Fejl ved {navn} - {path}: {e}")
    return None
```
